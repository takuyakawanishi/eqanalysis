```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats
import sys
sys.path.append("../..")
import src.eqa_takuyakawanishi.eqa as eqa


dir_data = "../../data/stationwise_fine_old_until_2019/"
file2read = "../../intermediates/organized_codes_pre_02.csv"
meta = pd.read_csv(file2read)
meta = meta.reset_index(drop=True)

df = pd.DataFrame(columns=["code", "address", "highest", "highestplus1", "duration", "probability_not_occurr"])

codes = meta["code_prime"]
set_dict = {"set_from": "1919-01-01", "set_to": "2020-12-31"}
count = 0
for i_code, code in enumerate(codes):
    print(code)
    frequency, regression, summary = \
        eqa.find_intensity_frequency_regression_summarize(
            meta, code, set_dict, dir_data=dir_data
        )
    address = meta.at[i_code, "address"]
    print(address)
    """
    try:
        frequency, regression, summary = \
            eqa.find_intensity_frequency_regression_summarize(
                meta, code, set_dict, dir_data=dir_data
            )
    except Exception as ex:
        print(code, ex)
    """
    if summary["duration"] > 10 * 365.2425:
        if summary["to"] == set_dict["set_to"]:
            # print("Station {} satisfies the condition.".format(code))
            highest = len(frequency)
            est = highest + 1
            duration = summary["duration"]
            if regression is not None:
                est_freq = np.exp(
                    - 10 ** (regression.intercept + regression.slope * est) * \
                    (duration / 365.2425)
                )
                # print(code, est, est_freq)
                df.loc[count, ["code", "highest", "highestplus1"]] = [code, highest, est]
                df.loc[count, ["duration", "probability_not_occurr", "address"]] = [duration, est_freq, address]
            count += 1

print(count)
print(len(meta))

```
